# Simple plotting in Python 

In order to visualize data, one needs to be able to plot it. Here I will show how to do basic plotting in python. Most basic plotting in python can be done using the matplotlib library. There is really good documentation for it on the malplotlib website itself, what I will provide here is a basic intro for people to start of with, the rest of the instructions one can follow from the matplotlib webiste

The matplotlib website is - https://matplotlib.org/examples/index.html

```python
%matplotlib inline 
import random as rand
import numpy as np 
import matplotlib.pyplot as plt 
```

I am going to generate some random numbers between 1 and 100 from a uniform distribution, for this I am using the 'random' library. I am also importing numpy and most importantly i am importing matplotlib as plt. 

```python
x = np.linspace(0,100,101) # this generates an array with numbers between 0 and 100. 
slope = 1.0 
b = 2 

y = (slope*x)+ b  # just as simple straight line  

# this is a line plot where the color of the line is red, how can you change the color of the line ? 
plt.figure()  # this line opens a new figure 
plt.plot(x,y,'-r') # plotting x and y with a red line line, the dash before the line indicates a connected line
plt.show()  # this line is needed to display the plot

```

```python
plt.figure()   
plt.plot(x,y,'-r', Linewidth = 20)  # changing the line width   
plt.show()
```

```python
x = np.linspace(0,100,21)  # reduced the number of points so that you can see the scatter plot

slope = 1.0 
b = 2 

y = (slope*x)+ b 

plt.figure()
plt.scatter(x,y)  #similar to the lines above but rather than a line plot we are plotting scatter plot
plt.show()


```

```python
# this part we add random noise to a straight line and plot a scatter plot
x = np.linspace(0,100,21)
random_number =  np.random.rand(21)  # generates N random numbers between 0 and 1 from a uniform distribution  
slope = 1 

y = (slope*x)+ 20*random_number 


plt.figure()
plt.scatter(x,y)  #similar to the lines above but rather than a line plot we are plotting scatter plot
plt.show()
```

So far we have only dealt with single plots, how is it that we use plot multiple plots on the same canvas. This is done using a for loop and showing the plot only after the for loop has run out

```python
x = np.linspace(0,100,101) # this generates an array with numbers between 0 and 100. 
slope = 1.0 
b =2

plt.figure() 
for i in range(-10,10):
    y = (i*x)+ b  # just as simple straight line  
    plt.plot(x,y)

plt.show() 
```
